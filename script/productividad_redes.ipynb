```python
# Importar librerias
import pandas as pd
```

```python
# Importar datos

vab_pb = pd.read_excel("serie_cgi_10_24.xls", sheet_name="VAB_pb")
puestos = pd.read_excel("serie_cgi_10_24.xls", sheet_name="Puestos")
horas = pd.read_excel("serie_cgi_10_24.xls", sheet_name="Horas")

# Filtrar datos

vab_pb = vab_pb.iloc[9:29] #filtrar filas
vab_pb = vab_pb.dropna(axis=1, how='all') #eliminar columnas vacias
puestos = puestos[9:29]
puestos = puestos.dropna(axis=1, how='all')
horas = horas[9:29]
horas = horas.dropna(axis=1, how='all')

# Lista del nuevo encabezado
nuevas_columnas = [
    "COD",
    "Sector de actividad económica",
    "2016 Q1", "2016 Q2", "2016 Q3", "2016 Q4", "2016 TOT",
    "2017 Q1", "2017 Q2", "2017 Q3", "2017 Q4", "2017 TOT",
    "2018 Q1", "2018 Q2", "2018 Q3", "2018 Q4", "2018 TOT",
    "2019 Q1", "2019 Q2", "2019 Q3", "2019 Q4", "2019 TOT",
    "2020 Q1", "2020 Q2", "2020 Q3", "2020 Q4", "2020 TOT",
    "2021 Q1", "2021 Q2", "2021 Q3", "2021 Q4", "2021 TOT",
    "2022 Q1", "2022 Q2", "2022 Q3", "2022 Q4", "2022 TOT",
    "2023 Q1", "2023 Q2", "2023 Q3", "2023 Q4", "2023 TOT",
    "2024 Q1", "2024 Q2"
]

vab_pb.columns = nuevas_columnas #asignar nuevo encabezado
puestos.columns = nuevas_columnas
horas.columns = nuevas_columnas

vab_pb
```

```python
# Crear el DataFrame de productividad dividiendo VAB_pb por Puestos
productividad_sector = vab_pb.iloc[:, 2:].divide(puestos.iloc[:, 2:])

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
productividad_sector.insert(0, "Sector de actividad económica", vab_pb["Sector de actividad económica"])
productividad_sector.insert(0, "COD", vab_pb["COD"])

productividad_sector
```

```python
# Crear el DataFrame de productividad dividiendo VAB_pb por Puestos
productividad_sector_horas = vab_pb.iloc[:, 2:].divide(horas.iloc[:, 2:])

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
productividad_sector_horas.insert(0, "Sector de actividad económica", vab_pb["Sector de actividad económica"])
productividad_sector_horas.insert(0, "COD", vab_pb["COD"])

productividad_sector_horas
```

```python
# Definir la función para calcular horas con distracción
def agregar_minutos(horas_df, minutos_extra_por_18):
    # Convertir las horas originales a minutos
    horas_en_minutos = horas_df.iloc[:, 2:] * 60  # Desde la tercera columna en adelante

    # Calcular los minutos extra (1 o 3 minutos cada 18 minutos)
    minutos_extra = (horas_en_minutos // 18) * minutos_extra_por_18

    # Sumar los minutos extra a los minutos originales
    horas_con_distraccion = horas_en_minutos + minutos_extra

    # Convertir de nuevo a horas
    return horas_con_distraccion / 60

# Calcular horas con distracción (1 minuto cada 18 minutos)
horas_distraccion_1m = agregar_minutos(horas, 1)

# Calcular horas con distracción (3 minutos cada 18 minutos)
horas_distraccion_3m = agregar_minutos(horas, 3)

# Crear DataFrame para horas con 1 minuto extra
horas_distraccion_1m.insert(0, "Sector de actividad económica", horas["Sector de actividad económica"])
horas_distraccion_1m.insert(0, "COD", horas["COD"])
horas_distraccion_1m = pd.DataFrame(horas_distraccion_1m, columns=horas.columns)

# Crear DataFrame para horas con 3 minutos extra
horas_distraccion_3m.insert(0, "Sector de actividad económica", horas["Sector de actividad económica"])
horas_distraccion_3m.insert(0, "COD", horas["COD"])
horas_distraccion_3m = pd.DataFrame(horas_distraccion_3m, columns=horas.columns)
```

```python
# Crear el DataFrame de productividad dividiendo VAB_pb por Puestos
productividad_sector_horas_distraccion_1m = vab_pb.iloc[:, 2:].divide(horas_distraccion_1m.iloc[:, 2:])

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
productividad_sector_horas_distraccion_1m.insert(0, "Sector de actividad económica", vab_pb["Sector de actividad económica"])
productividad_sector_horas_distraccion_1m.insert(0, "COD", vab_pb["COD"])

productividad_sector_horas_distraccion_1m
```

```python
# Crear el DataFrame de productividad dividiendo VAB_pb por Puestos
productividad_sector_horas_distraccion_3m = vab_pb.iloc[:, 2:].divide(horas_distraccion_3m.iloc[:, 2:])

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
productividad_sector_horas_distraccion_3m.insert(0, "Sector de actividad económica", vab_pb["Sector de actividad económica"])
productividad_sector_horas_distraccion_3m.insert(0, "COD", vab_pb["COD"])

productividad_sector_horas_distraccion_3m
```

```python
# Generar merma_productividad_1m
merma_productividad_1m = productividad_sector_horas.iloc[:, 2:] - productividad_sector_horas_distraccion_1m.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
merma_productividad_1m.insert(0, "Sector de actividad económica", productividad_sector_horas["Sector de actividad económica"])
merma_productividad_1m.insert(0, "COD", productividad_sector_horas["COD"])

# Generar merma_productividad_3m
merma_productividad_3m = productividad_sector_horas.iloc[:, 2:] - productividad_sector_horas_distraccion_3m.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
merma_productividad_3m.insert(0, "Sector de actividad económica", productividad_sector_horas["Sector de actividad económica"])
merma_productividad_3m.insert(0, "COD", productividad_sector_horas["COD"])

# Verificar los resultados
print("Merma Productividad 1m:")
print(merma_productividad_1m.head())

print("\nMerma Productividad 3m:")
print(merma_productividad_3m.head())
```

```python
# Calcular valor_merma_productividad_1m
valor_merma_productividad_1m = merma_productividad_1m.iloc[:, 2:] * vab_pb.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
valor_merma_productividad_1m.insert(0, "Sector de actividad económica", merma_productividad_1m["Sector de actividad económica"])
valor_merma_productividad_1m.insert(0, "COD", merma_productividad_1m["COD"])

# Calcular valor_merma_productividad_3m
valor_merma_productividad_3m = merma_productividad_3m.iloc[:, 2:] * vab_pb.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
valor_merma_productividad_3m.insert(0, "Sector de actividad económica", merma_productividad_3m["Sector de actividad económica"])
valor_merma_productividad_3m.insert(0, "COD", merma_productividad_3m["COD"])

# Verificar los resultados
print("Valor Merma Productividad 1m:")
print(valor_merma_productividad_1m.head())

print("\nValor Merma Productividad 3m:")
print(valor_merma_productividad_3m.head())
```

```python
# Calcular %_merma_productividad_1m
porcentaje_merma_productividad_1m = valor_merma_productividad_1m.iloc[:, 2:] / vab_pb.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
porcentaje_merma_productividad_1m.insert(0, "Sector de actividad económica", valor_merma_productividad_1m["Sector de actividad económica"])
porcentaje_merma_productividad_1m.insert(0, "COD", valor_merma_productividad_1m["COD"])

# Calcular %_merma_productividad_3m
porcentaje_merma_productividad_3m = valor_merma_productividad_3m.iloc[:, 2:] / vab_pb.iloc[:, 2:]

# Añadir las columnas "COD" y "Sector de actividad económica" al nuevo DataFrame
porcentaje_merma_productividad_3m.insert(0, "Sector de actividad económica", valor_merma_productividad_3m["Sector de actividad económica"])
porcentaje_merma_productividad_3m.insert(0, "COD", valor_merma_productividad_3m["COD"])

# Verificar los resultados
print("Porcentaje Merma Productividad 1m:")
print(porcentaje_merma_productividad_1m.head())

print("\nPorcentaje Merma Productividad 3m:")
print(porcentaje_merma_productividad_3m.head())
```

```python
# Nombre del archivo a guardar
nombre_archivo = "analisis_productividad.xlsx"

# Exportar los DataFrames a un archivo Excel
with pd.ExcelWriter(nombre_archivo, engine="openpyxl") as writer:
    vab_pb.to_excel(writer, sheet_name="VAB_pb", index=False)
    productividad_sector.to_excel(writer, sheet_name="Productividad_Sector", index=False)
    productividad_sector_horas.to_excel(writer, sheet_name="Prod_Sector_Horas", index=False)
    merma_productividad_1m.to_excel(writer, sheet_name="Merma_Prod_1m", index=False)
    merma_productividad_3m.to_excel(writer, sheet_name="Merma_Prod_3m", index=False)
    valor_merma_productividad_1m.to_excel(writer, sheet_name="Valor_Merma_1m", index=False)
    valor_merma_productividad_3m.to_excel(writer, sheet_name="Valor_Merma_3m", index=False)
    porcentaje_merma_productividad_1m.to_excel(writer, sheet_name="Porc_Merma_1m", index=False)
    porcentaje_merma_productividad_3m.to_excel(writer, sheet_name="Porc_Merma_3m", index=False)
```
